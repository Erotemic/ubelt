# Ubelt's cmd method

```python
import ubelt as ub
```

```python
info = ub.cmd(['python', '--version'], shell=False)
type(info)
print(info.keys())
print(ub.repr2(info, sort=0))
```

```python
info = ub.cmd(['python', '--verssion'])
print(info['ret'])
```

```python
print(ub.cmd.__doc__)
```

```python
ub.cmd('echo hello world && ping localhost', verbose=3, shell=True)
```
